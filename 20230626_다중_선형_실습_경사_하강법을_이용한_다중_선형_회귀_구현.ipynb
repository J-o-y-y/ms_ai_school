```python
# 다중 선형 실습 : 경사 하강법을 이용한 다중 선형 회귀 구현
```

```python
import numpy as np
```

```python
# 데이터 생성
```

```python
x1 = np.array([1,2,3,4,5])
x2 = np.array([0,1,0,1,0])

y= np.array([3,5,7,9,11])
```

```python
def gradient_descent(x1, x2, y, lr, iterations) :

  # 초기값 설졍
    n = len(y)
    beta_0 = 0
    beta_1 = 0
    beta_2 = 0

  # 경사하강법 수행
    for i in range(iterations) :
        y_pred = beta_0 + beta_1 * x1 + beta_2 * x2

        error = y - y_pred - y

        # 업데이트
        beta_0 -= lr * (1/n) * np.sum(error)
        beta_1 -= lr * (1/n) * np.sum(error * x1)
        beta_2 -= lr * (1/n) * np.sum(error * x2)

    return beta_0, beta_1, beta_2
```

```python
beta_0, beta_1, beta_2 = gradient_descent (x1, x2, y, 0.1, 1000)

print("beta_0 >>", beta_0)
print("beta_1 >>", beta_1)
print("beta_2 >>", beta_2)
```

```python
x1_new = 2
x2_new = 1

# 예측되어야 하는 값 3

y_pred_temp = beta_0 + beta_1 * x1_new + beta_2 * x2_new

print("Predicted y value >> ", y_pred_temp)
```
