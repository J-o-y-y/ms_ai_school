1. 먼저 가상의 데이터를 생성해보겠습니다. 이 데이터는 1학년부터 3학년까지의 학생들의 수학, 영어, 국
어 점수를 나타내며, 각 학년당 100명씩의 학생이 있다고 가정합니다.

```python
# 필요한 라이브러리 불러오기
import pandas as pd
import numpy as np

# 랜덤 시드 설정 (이를 통해 결과를 재현할 수 있습니다)
np.random.seed(0)

# 가상의 점수 데이터 생성 (0점부터 100점까지)
math_scores = np.random.randint(0, 101, 300) 
english_scores = np.random.randint(0, 101, 300)
korean_scores = np.random.randint(0, 101, 300)

# 학년 데이터 생성
grades = np.repeat([1, 2, 3], 100)

# 데이터프레임 생성
df = pd.DataFrame({
    'Grade': grades,
    'Math': math_scores,
    'English': english_scores,
    'Korean': korean_scores
})

# 상관곱 실행
correlation_matrix = df.corr()

# 결과 출력
print(correlation_matrix)
```

3. 상관 관계 분석 – 히트맵 출력

```python
# 필요한 라이브러리 불러오기
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 랜덤 시드 설정
np.random.seed(0)

# 가상의 점수 데이터 생성 (0점부터 100점까지)
math_scores = np.random.randint(0, 101, 300)
english_scores = np.random.randint(0, 101, 300)
korean_scores = np.random.randint(0, 101, 300)

# 학년 데이터 생성
grades = np.repeat([1, 2, 3], 100)

# 데이터프레임 생성
df = pd.DataFrame({
    'Grade': grades,
    'Math': math_scores,
    'English': english_scores,
    'Korean': korean_scores
})

# 상관곱 계산
correlation_matrix = df.corr()

# 히트맵 생성
plt.figure(figsize=(10, 8)) # 히트맵 크기 설정
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1) # annot 옵션으로 상관계수값 출력, cmap으로 색상 팔레트 설정, vmin과 vmax로 색상 팔레트 범위 설정
plt.title('Correlation heatmap')
plt.show()
```

4. 과목별로 성적 분포와 통계치를 분석 판다스의 describe() 함수를 사용

```python
# pandas의 describe 메소드 사용하여 각 열의 통계치 계산
statistics = df[['Math', 'English', 'Korean']].describe()

# 결과 출력
print(statistics)
```

5. 과목별 통계치 시각화 – 히스토그램

```python

# 랜덤 시드 설정
np.random.seed(0)

# 가상의 점수 데이터 생성 (0점부터 100점까지)
math_scores = np.random.randint(0, 101, 300)
english_scores = np.random.randint(0, 101, 300)
korean_scores = np.random.randint(0, 101, 300)

# 학년 데이터 생성
grades = np.repeat([1, 2, 3], 100)

# 데이터프레임 생성
df = pd.DataFrame({
    'Grade': grades,
    'Math': math_scores,
    'English': english_scores,
    'Korean': korean_scores
})

# 히스토그램 그리기
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(df['Math'], bins=20, color='blue', alpha=0.7)
plt.title('Math Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.hist(df['English'], bins=20, color='red', alpha=0.7)
plt.title('English Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
plt.hist(df['Korean'], bins=20, color='green', alpha=0.7)
plt.title('Korean Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')

plt.tight_layout() # 서브플롯 간의 간격을 자동으로 조정
plt.show()
```
